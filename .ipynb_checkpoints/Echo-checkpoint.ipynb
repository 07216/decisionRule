```python
import rALP
import decisionRule

solver = rALP.rALP()
solver.readIn('rm_200_4_1.0_4.0.txt')
solver.product()
solver.leg()
solver.construct()
#solver.echoVal()
#solver.addVAR()
#solver.addOpt()
```

```python
decisionSolver = decisionRule.decisionRule(solver)
decisionSolver.construct()
decisionSolver.echoInput()
#decisionSolver.echoVal()
decisionSolver.addVar()
decisionSolver.addOpt()
decisionSolver.addConstr()
#decisionSolver.writeMPS()
#decisionSolver.echoOpt()
```

```python
decisionSolver.solve()
```
